```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../summary_data.csv')
```

```python
colours4 = ['#26547c', '#ef476f','#06d6a0','#70d6ff']

fig, ax = plt.subplots(figsize=(8,8))
sns.pointplot(data=data, x='vibration', y='error', hue='vibration', ax=ax,
              errorbar=None, marker='_', markersize=20, linewidth=4, legend=True,
              order=['No Vibration', 'Biceps', 'Dual', 'Triceps'], palette=colours4)
sns.swarmplot(data=data, x='vibration', y='error', hue='vibration', 
              ax=ax, alpha=0.5, order=['No Vibration', 'Biceps', 'Dual', 'Triceps'],
              palette=colours4)
ax.axhline(0,linestyle='--', color='black')
ax.legend(frameon=False)
sns.despine()
ax.set_ylabel("Error ($\circ$)")
ax.set_xlabel("")
fig.savefig('../results/mean_end_point.svg')
plt.show()
```

```python
colours3 = ['#ef476f','#06d6a0','#70d6ff']

fig, ax = plt.subplots(figsize=(8,8))
error_data = data[data['vibration'] != "No Vibration"]
sns.pointplot(data=error_data, x='vibration', y='elbow_end_deg_diff', hue='vibration', ax=ax,
              errorbar=None, marker='_', markersize=20, linewidth=4, legend=True,
              order=['Biceps', 'Dual', 'Triceps'], palette=colours3)
sns.swarmplot(data=error_data, x='vibration', y='elbow_end_deg_diff', hue='vibration', 
              ax=ax, alpha=0.5, order=['Biceps', 'Dual', 'Triceps'], palette=colours3)
ax.axhline(0,linestyle='--', color='black')
ax.legend(frameon=False)
ax.set_ylabel("Error ($\circ$)")
ax.set_xlabel("")
sns.despine()
fig.savefig('../results/mean_vibration_error.svg')
plt.show()
```
