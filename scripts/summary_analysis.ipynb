```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../summary_data.csv')
```

```python
colours4 = ['#26547c', '#ef476f','#06d6a0','#70d6ff']

fig, ax = plt.subplots(1,3, figsize=(16,6))
sns.pointplot(data=data, x='vibration', y='error', hue='vibration', ax=ax[0],
              errorbar=None, marker='_', markersize=20, linewidth=4, legend=True,
              order=['No Vibration', 'Biceps', 'Dual', 'Triceps'], palette=colours4)
sns.swarmplot(data=data, x='vibration', y='error', hue='vibration', 
              ax=ax[0], alpha=0.7, order=['No Vibration', 'Biceps', 'Dual', 'Triceps'],
              palette=colours4)
ax[0].axhline(0,linestyle='--', color='black')
ax[0].legend(frameon=False)
sns.despine(offset=10)
ax[0].set_ylabel("Error ($\circ$)")
ax[0].set_xlabel("")
ax[0].set_title("Target Constant Error")
# ax[0].set_ylim(-12,8)
# ax[0].set_yticks([])
# fig.savefig('../results/mean_end_point.svg')
# plt.show()
colours3 = ['#ef476f','#06d6a0','#70d6ff']

# fig, ax = plt.subplots(figsize=(8,8))
error_data = data[data['vibration'] != "No Vibration"]
sns.pointplot(data=error_data, x='vibration', y='elbow_end_deg_diff', hue='vibration', ax=ax[1],
              errorbar=None, marker='_', markersize=20, linewidth=4, legend=False,
              order=['Biceps', 'Dual', 'Triceps'], palette=colours3)
sns.swarmplot(data=error_data, x='vibration', y='elbow_end_deg_diff', hue='vibration', 
              ax=ax[1], alpha=0.7, order=['Biceps', 'Dual', 'Triceps'], palette=colours3)
ax[1].axhline(0,linestyle='--', color='black')
# ax[1].legend(frameon=False)
ax[1].set_ylabel("Error ($\circ$)")
ax[1].set_xlabel("")
sns.despine(offset=10)
ax[1].set_title("Vibration Error")
# fig.savefig('../results/mean_vibration_error.svg')
# plt.show()

data_std = pd.read_csv('../summary_data_std.csv')
data_std.head(4)


# fig, ax = plt.subplots(figsize=(8,8))
sns.pointplot(data=data_std, x='vibration', y='error', hue='vibration', ax=ax[2],
              errorbar=None, marker='_', markersize=20, linewidth=4, legend=False,
              order=['No Vibration', 'Biceps', 'Dual', 'Triceps'], palette=colours4)
sns.swarmplot(data=data_std, x='vibration', y='error', hue='vibration', 
              ax=ax[2], alpha=0.7, order=['No Vibration', 'Biceps', 'Dual', 'Triceps'], palette=colours4)
# ax[2].legend(frameon=False)
sns.despine(offset=10)
ax[2].set_ylabel("Error ($\circ$)")
ax[2].set_xlabel("")
ax[2].set_ylim(0,10)
ax[2].set_title("Target Variable Error")
plt.tight_layout()
plt.show()
```
